Question 7
   If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
  If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Create order Frequency
orders_products_merged['median_orders'] = orders_products_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
orders_products_merged.loc[orders_products_merged['median_orders'] >20, 'order_frequency_flag'] = 'Non-frequent_customer'
```

```python
orders_products_merged.loc[orders_products_merged['median_orders'] >10 & (orders_products_merged['median_orders']<=20), 'order_frequency_flag'] = 'Regular_customer'
```

```python
orders_products_merged.loc[orders_products_merged['median_orders'] <=10, 'order_frequency_flag'] = 'frequent_customer'
```

```python
orders_products_merged['order_frequency_flag'].value_counts(dropna = False)
```
